```python
# LOADING THE DATASET
import h5py

data = h5py.File("ydata-tvsum50.mat", "r")
dataset = data['tvsum50']
print(dataset.keys())
```

```python
# ----------------------------
# GET SCORES
# ---------------------------
# 
import numpy as np
scores_per_video = []

for i in range(dataset['user_anno'].shape[0]):
    ref = dataset['user_anno'][i, 0]
    arr = data[ref][()]
    scores_per_video.append(arr)

# average scores
avg_scores_per_video = [
    np.mean(s, axis=0) for s in scores_per_video
]
```

```python
video_ids = []

for i in range(dataset['video'].shape[0]):
    ref = dataset['video'][ i,0]
    arr = data[ref][()]

    # flatten safely
    arr = arr.flatten()

    vid = ''.join(chr(int(c)) for c in arr)
    video_ids.append(vid)

print(video_ids[:10])
```

```python
video_to_scores = {
    video_ids[i]: avg_scores_per_video[i]
    for i in range(len(video_ids))
}
```

```python
# titles
titles = []
for i in range(dataset['title'].shape[0]):
    ref = dataset['title'][i,0]
    arr = data[ref][()].flatten()
    title = ''.join(chr(int(c)) for c in arr)
    titles.append(title)

print(titles[:5])
```

```python
# convert titles to youtube links 
import yt_dlp
import time

def search_youtube(query):
    try:
        with yt_dlp.YoutubeDL({'quiet': True}) as ydl:
            result = ydl.extract_info(f"ytsearch1:{query}", download=False)
            return result['entries'][0]['webpage_url']
    except:
        return None

links = []

for title in titles:
    link = search_youtube(title)
    links.append(link)
    print(f"{title} → {link}")
    
    time.sleep(1)  # avoid rate limiting
```

```python
import os
import json
import numpy as np
import re


def extract_video_id(name):
    parts = name.split()
    last = parts[-1]

    # remove weird chars like "-" or "."
    last = re.sub(r"[^\w\-]", "", last)

    return last


def get_score_for_shot(start, end, frame_scores, fps=30):
    start_frame = int(start * fps)
    end_frame = int(end * fps)

    start_frame = max(0, start_frame)
    end_frame = min(len(frame_scores), end_frame)

    if end_frame <= start_frame:
        return 0.0

    return float(np.mean(frame_scores[start_frame:end_frame]))
```

```python
# saving links to a file
with open("links.txt", "w") as f:
    for link in links:
        if link:
            f.write(link + "\n")
```

Cleaning Titles

```python
import os
import json
import numpy as np
import re

def clean_title(text):
    text = text.lower()
    text = re.sub(r"[^\w\s]", "", text)
    text = re.sub(r"\s+", " ", text)
    return text.strip()
```

Extract Scores

```python
# ----------------------------
# GET SCORES
# ----------------------------
scores_per_video = []

for i in range(dataset['user_anno'].shape[0]):
    ref = dataset['user_anno'][i, 0]
    arr = data[ref][()]   # shape: [num_users, num_frames]
    scores_per_video.append(arr)


# average across users
avg_scores_per_video = [
    np.mean(s, axis=0) for s in scores_per_video
]
```

Titles to Scores Map

```python
title_to_scores = {}

for i in range(len(titles)):
    clean_t = clean_title(titles[i])
    title_to_scores[clean_t] = avg_scores_per_video[i]
```

Getting shot Scores

```python
def get_score_for_shot(start, end, frame_scores, fps=30):
    start_frame = int(start * fps)
    end_frame = int(end * fps)

    start_frame = max(0, start_frame)
    end_frame = min(len(frame_scores), end_frame)

    if end_frame <= start_frame:
        return 0.0

    return float(np.mean(frame_scores[start_frame:end_frame]))
```

# Adding scores to the merged json

```python
MERGED_DIR = "data/merged"
OUTPUT_DIR = "data/final_with_scores"

os.makedirs(OUTPUT_DIR, exist_ok=True)

for file in os.listdir(MERGED_DIR):

    if not file.endswith(".json"):
        continue

    raw_name = os.path.splitext(file)[0]

    # extract YouTube ID
    video_id = extract_video_id(raw_name)

    print(f"Processing: {video_id}, {raw_name}")
    print(video_id in video_to_scores)

    if video_id not in video_to_scores:
        print(f"⚠️ No match for: {raw_name}")
        continue

    frame_scores = video_to_scores[video_id]

    input_path = os.path.join(MERGED_DIR, file)

    with open(input_path, "r") as f:
        shots = json.load(f)

    # ----------------------------
    # STEP 1: compute raw scores
    # ----------------------------
    raw_scores = []

    for shot in shots:
        start = shot["start"]
        end = shot["end"]

        score = get_score_for_shot(start, end, frame_scores)
        raw_scores.append(score)

    # ----------------------------
    # STEP 2: normalize per video
    # ----------------------------
    raw_scores = np.array(raw_scores)

    if raw_scores.max() != raw_scores.min():
        norm_scores = (raw_scores - raw_scores.min()) / (raw_scores.max() - raw_scores.min())
    else:
        norm_scores = raw_scores

    # ----------------------------
    # STEP 3: assign back
    # ----------------------------
    for i, shot in enumerate(shots):
        shot["score"] = float(norm_scores[i])

    # ----------------------------
    # SAVE
    # ----------------------------
    output_path = os.path.join(OUTPUT_DIR, file)

    with open(output_path, "w") as f:
        json.dump(shots, f, indent=2)

    print(f"✅ Saved: {output_path}")
```
