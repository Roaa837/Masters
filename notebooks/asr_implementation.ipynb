```python
def run_asr(video_path, output_json):
    import whisper, json, os

    model = whisper.load_model("base")
    result = model.transcribe(video_path, verbose=False,
    fp16=False, no_speech_threshold=0.6,
    logprob_threshold=-1.0)

    os.makedirs(os.path.dirname(output_json), exist_ok=True)

    with open(output_json, "w", encoding="utf-8") as f:
        json.dump(result["segments"], f, indent=2)

    return result["segments"]
```

## Batch Runner


```python
import os
import whisper

RAW_VIDEO_DIR = "../data/raw_videos"
ASR_DIR = "../data/asr"

os.makedirs(ASR_DIR, exist_ok=True)

# Load Whisper ONCE
model = whisper.load_model("base")

for video in os.listdir(RAW_VIDEO_DIR):
    if not video.lower().endswith(".mp4"):
        continue

    video_path = os.path.join(RAW_VIDEO_DIR, video)
    asr_path = os.path.join(ASR_DIR, video.replace(".mp4", ".json"))

    if os.path.exists(asr_path):
        print(f"[SKIP] ASR already exists for {video}")
        continue

    print(f"[ASR] Processing {video}")
    run_asr(video_path, asr_path)

print("✅ ASR extraction complete")
```
